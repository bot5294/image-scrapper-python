```python
query = "aqib ahmed"
response = requests.get("https://www.google.com/search?q={query}&tbm=isch&ved=2ahUKEwjo0NvVnp6FAxUmT2wGHeQVDGYQ2-cCegQIABAA&oq=aqib+ahmed&gs_lp=EgNpbWciCmFxaWIgYWhtZWRIAFAAWABwAHgAkAEAmAEAoAEAqgEAuAEMyAEAigILZ3dzLXdpei1pbWc&sclient=img&ei=zTMJZqi2MKaeseMP5KuwsAY&bih=945&biw=1131&prmd=ivnsmbtz&rlz=1C1RXQR_enIN1047IN1047")
```

```python
soup = BeautifulSoup(response.content,"html.parser")
```

```python
img_tags = soup.find_all("img")
```

```python
del img_tags[0]
```

```python
len(img_tags)
```

```python
img_data_mongo = []
for i in img_tags:
    img_url = i['src']
    img_data = requests.get(img_url).content
    mydict = {"index":img_url,"image":img_data}
    img_data_mongo.append(mydict)
    with open(os.path.join(save_dir,f"{query}_{img_tags.index(i)}.jpg"),"wb") as f :
                           f.write(img_data)
```
